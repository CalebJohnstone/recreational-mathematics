```python
import sympy

original_c = c = 41
limit = 100

prime_percentages = []
moving_prime_percentages = []

for i in range(1, limit+1):
    prime_count = 0
    moving_prime_count = 0
    found_below_100 = False

    for j in range(1, i+1):
        if sympy.isprime(j**2 - j + original_c):
            prime_count += 1
        if sympy.isprime(j**2 + j + c):
            moving_prime_count += 1

    prime_percentage = prime_count / i * 100
    prime_percentages.append(prime_percentage)

    moving_prime_percentage = moving_prime_count / i * 100
    moving_prime_percentages.append(moving_prime_percentage)

    if not found_below_100 and moving_prime_percentage < 100:
        found_below_100 = True

        c += 2
        while not sympy.isprime(c):
            c += 2

# plot the prime percentages
import matplotlib.pyplot as plt
plt.plot(list(range(1,limit+1)), prime_percentages)
plt.plot(list(range(1,limit+1)), moving_prime_percentages)
plt.ylabel('Prime Percentage')
plt.xlabel('n')
plt.show()
```
